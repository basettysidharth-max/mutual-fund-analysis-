```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd

for file in uploaded.keys():
    print("\n" + "="*70)
    print("FILE:", file)

    df = pd.read_csv(file)

    print("\nShape:")
    print(df.shape)

    print("\nData Types:")
    print(df.dtypes)

    print("\nFirst 5 Rows:")
    print(df.head())

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nDuplicate Rows:")
    print(df.duplicated().sum())
```
